```python
import pandas as pd
```

```python
#Here we are creating a DataFrame from a directory
df = pd.DataFrame(
    {
    'name': ['Adam', 'Bob', 'Dave', 'Fred'],
    'age': [15, 16, 16, 15],
    'test1': [95, 81, 89, None],
    'test2': [80, 82, 84, 88],
    'teacher': ['Ashby', 'Ashby', 'Jones', 'Jones']
    }
)

df
```

```python
#For the next method, we will need to first define three Pandas series
purchase_1 = pd.Series(
    {
        'Cost': 22.5,
        'Item Purchased': 'Dog food',
        'Name': 'Chris'
    }
) 
purchase_2 = pd.Series(
    {
        'Cost': 2.5,
        'Item Purchased': 'Kitty Litter',
        'Name': 'Kevin'
    }
) 
purchase_3 = pd.Series(
    {
        'Cost': 5,
        'Item Purchased': 'Bird Seed',
        'Name': 'Vinod'
    }
) 
```

```python
#Here we are combining the 3 series into a data frame
df2 = pd.DataFrame([purchase_1, purchase_2, purchase_3],
            index=['Store 1', 'Store 1', 'Store 2'])

df2
```
