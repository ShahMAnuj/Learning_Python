```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'column 1': [1,2,3,4,5,6,7],
        'column 2': [2,4,7,2,5,4,3],
        'column 3': [9,10,11,12,13,14,15]
    }
)
df
```

```python
#Here, we count the number of times each value
#occurs in column 2 of our DataFrame
df['column 2'].value_counts()
```

```python
#Here, we count the numbe of times each value
#occurs in row 3 of our DataFrame
df.loc[3].value_counts()
```
