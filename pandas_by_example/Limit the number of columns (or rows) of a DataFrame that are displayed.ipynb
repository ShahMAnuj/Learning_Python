```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns", 4)
```

```python
df = pd.DataFrame(np.array([[6, 2, 6,0, 7, 5, 2, 9], [3, 4, 9,0, 4, 7, 1, 7], [12, 4, 2,0, 5, 7, 9, 10], [7,9,4,0,9,4,3,1]]),
                   columns=['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight'])
```

```python
#Even though there are eight columns in our dataframe, 
#but because we have set display.max_columns to 4,
#thus, only 4 columns will be displayed
df
```

```python
pd.set_option("display.max_rows", 2)
```

```python
#Even though there are two rows in our dataframe,
#but because we have set display.max_rows to 2,
#thus, only 2 rows will be displayed
df
```
