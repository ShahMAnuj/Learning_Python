```python
import scipy.integrate as integrate

def inner_product(f,g,a,b):
    product = lambda z : f(z) * g(z)
    P = b - a
    answer = (2/P) * integrate.quad(product,a,b)[0]
    return answer
```

```python
import numpy as np
import math

def fourier_coefficients(f,a,b,N):
    A = np.zeros((1,N))
    B = np.zeros((1,N))
    
    #computing the A(0) coefficient
    g = lambda t : 1
    A0 = inner_product(f,g,a,b)
    if A0 < 10**(-5):
        A0 = 0
    
    #computing the A(k) coefficients
    A = np.zeros(N)
    for k in range(N):
        g = lambda t : math.cos(k*t)
        A[k] = inner_product(f,g,a,b)
        if A[k] < 10**(-5):
            A[k] = 0
    
    #computing the B(k) coefficients
    B = np.zeros(N)
    for k in range(N):
        g = lambda t : math.sin(k*t)
        B[k] = inner_product(f,g,a,b)
        if B[k] < 10**(-5):
            B[k] = 0
            
    return [A0, A, B]       
```

### Fourier series of sin(40t) ###

```python
f = lambda t : math.sin(40*t)
answer = fourier_coefficients(f,-1,1,100)

import matplotlib.pyplot as plt
print("A(0) is {}".format(answer[0]))
fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].plot(answer[1])
axs[0].set(
    xlabel='k',
    ylabel='A(k)',
    title='cosine coefficients'
)
axs[1].plot(answer[2])
axs[1].set(
    xlabel='k',
    ylabel='B(k)',
    title='sine coefficients'
)
```
