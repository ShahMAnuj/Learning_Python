```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating an array which starts at x=1, ends at x=6, and has spacings of 0.1
x = np.arange(1, 6, 0.1)
#y = exp(1/x)
y = np.exp(np.reciprocal(x))
```

```python
#Plotting!
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set(
    xlabel='x',
    ylabel='exp(1/x)',
    title='essential singularity'
)
ax.grid()
plt.show()
```
